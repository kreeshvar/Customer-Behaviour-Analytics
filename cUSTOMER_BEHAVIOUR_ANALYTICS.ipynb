```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns=df.columns.str.lower()
df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
df
```

```python
labels=['Yound Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)

df[['age','age_group']].head(10)
```

```python
df.columns
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequncy_mapping={
    'Fortnightly':14,
    'Weekly':7,  
    'Annually':365, 
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}
df['purchase_frequncy_days']=df['frequency_of_purchases'].map(frequncy_mapping)
df[['purchase_frequncy_days','frequency_of_purchases']].head(10)
```
